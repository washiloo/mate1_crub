## Ejercicio 2

En este ejercicio se dan dos conjuntos $A$ y $B$ definidos por comprensión y se pide graficar la región del plano que representa el conjunto $C = A \times B$, es decir, el producto cartesiano entre $A$ y $B$. Además, se pide escribir las condiciones que debe cumplir cualquier par ordenado $(x,y)$ para pertenecer a dicho conjunto.

Lo _primero_ que vamos a hacer es establecer las condiciones que debe cumplir $(x,y)$ para pertenecer a $C$. Esta condición va a estar dada por una proposición lógica o una operación entre proposiciones.

Por ejemplo, si $A = \left\{x \in \mathbb{R} \text{ / } x > 2\right\}$ y $B = \left\{x \in \mathbb{R} \text{ / } x \geq 3\right\}$, tenemos que el producto cartesiano es:

$$C = A \times B = \left\{(x,y) \in \mathbb{R}^2 \text{ / } x > 2 \wedge y \geq 3\right\}$$

Esto es así porque, según la definición de producto cartesiano (¡repasar la teoría!) los valores de $x$ en el par ordenado $(x,y)$ (la primera coordenada) son los valores que toma $x$ en el conjunto $A$ (el primer conjunto en el producto cartesiano), mientras que los valores de $y$ en el par ordenado $(x,y)$ (la segunda coordenada) son los valores que toma $x$ en el conjunto $B$ (el segundo conjunto en el producto cartesiano).

Veamos cómo podemos graficar el conjunto $C$ en el plano:

```python
x_lim = [0,5] # Límites de la variable x para graficar [mínimo, máximo]
y_lim = [0,7] # Límites de la variable y para graficar [mínimo, máximo]

cond = '(x > 2) & (y >= 3)' # Defino la proposición que hacen verdadera los elementos del conjunto

graficar_region(x_lim,y_lim,cond) # Grafico la región (sin la frontera)
```

Como vemos, hay varias cosas nuevas en este código. Vayamos por partes:

- Las variables `x_lim` e `y_lim` son **listas** que contienen dos números que representan los valores extremos de cada variable en el gráfico que queremos construir. Las listas son objetos de _python_, y se definen usando _corchetes_. Podemos poner cualquier tipo de objeto dentro de una lista, y el orden en el que lo hacemos _sí importa_. En este caso, son sólo dos números. El primer número corresponde al valor más pequeño y el segundo número corresponde al valor más grande. En el ejemplo, `x_lim = [0,5]` significa que vamos a graficar el eje $x$ entre los valores 0 y 5. Algo análogo ocurre con la otra variable.   
- La línea `cond = '(x > 2) & (y >= 3)'` hace gran parte de la magia. Este pedazo de código define un _string_ con la condición que deben cumplir los elementos del conjunto que queremos graficar. En este caso, la condición es una _conjunción_ entre dos proposiciones. La primera proposicióm es $x > 2$ y la segunda proposición es $y \geq 3$. Para hacer la conjunción, usamos el símbolo **&** y colocamos las proposiciones individuales entre paréntesis (esto es para mayor claridad). Acá va una lista de los otros operadores lógicos que solemos usar:

  - Disyunción inclusiva ($\vee$), en _python_ es | (barra vertical, suele teclearse ALT + 1).
  - Disyunción exclusiva ($\veebar$), en _python_ es ^ (acento circunflejo).
  - Negación ($\sim$), en _python_ es ~ (virgulilla).  
  
  Estos operadores se usan de la misma manera que vimos en clase. 
- Finalmente, la línea `graficar_region(x_lim,y_lim,cond)` grafica la región que queremos. Internamente, el código define una grilla de pares ordenados usando los límites que le dijimos antes en las variables `x_lim` e `y_lim`. Por defecto, crea un millón de pares ordenados equiespaciados entre esos límites. En el ejemplo, como $x$ varía entre 0 y 5 e $y$ varía entre 0 y 7, dentro de esa grilla estarán los pares ordenados (0,0), (0,2.5), (1.4,3.7), (0,7), (5,0), (5,5), (5,7), entre muchos otros. Luego, la función toma todos estos pares ordenados y **evalúa si cumplen la condición** que le indicamos. Finalmente, sólo grafica los pares ordenados que cumplen la condición. _Voilà!_

Ahora bien, nuestro gráfico está incompleto, porque no graficamos la _frontera_ del conjunto. De manera informal, la frontera de un conjunto en $\mathbb{R}^2$ es el borde de la región que representa el conjunto. En nuestro ejemplo, la frontera está formada por dos semirrectas: la semirrecta vertical $x = 2$ para $y \geq 3$ y la semirrecta horizontal $y = 3$ para $x > 2$. Además, la semirrecta vertical _no pertenece al conjunto_, pues la desigualdad es estricta ($x > 2$), mientras que la semirrecta horizontal _sí pertenece al conjunto_, pues la expresión $y \geq 3$ incluye la posibilidad de que $y = 3$. Como vimos en clase, usamos _líneas cortadas_ para graficar fronteras que no pertenecen al conjunto, y _líneas llenas_ para graficar fronteras que sí pertenecen al conjunto. En la siguiente celda de código, graficamos el mismo conjunto que antes pero le agregamos la frontera (por partes).

```python
from tp_2_mod import *

x_lim = [0,5] # Límites de la variable x para graficar [mínimo, máximo]
y_lim = [0,7] # Límites de la variable y para graficar [mínimo, máximo]

cond = '(x > 2) & (y >= 3)' # Defino la proposición que hacen verdadera los elementos del conjunto

fr1 = frontera(x = '2',tipo = '--') # Defino una parte de la frontera del conjunto (semirrecta vertical)
fr2 = frontera(y = '3',tipo = '-') # Defino una parte de la frontera del conjunto (semirrecta horizontal)

graficar_region(x_lim,y_lim,cond,fronteras = [fr1,fr2]) # Grafico la región (con la frontera)
```

Las únicas líneas de código nuevas acá son:

- La línea `fr1 = frontera(x = '2',tipo = '--')` define la semirrecta vertical que pertenece a la frontera del conjunto. Usamos la función _frontera_ para definirla, y los argumentos que pasamos a la función son _x_ y _tipo_. 
  - En este caso, como la recta es vertical tenemos una función $x = g(y)$, es decir que acá _x es la variable dependiente_ (función) e _y es la variable independiente_. Entonces, pasamos la función como un _string_ (entre comillas) en el argumento _x_. En este caso, la función es constante e igual a 2, por lo que $x = g(y) = 2$ y entonces escribimos _x = '2'_ en el argumento de la función _frontera_. Los límites de  $y$ son detectados automáticamente. Si quisiéramos hacer una función diferente, bastaría con escribir el código necesario. Por ahora, sólo trabajaremos con fronteras lineales, así que no te enrosques con esto (más abajo hay otro ejemplo más complejo).   
  - Como esta parte de la frontera no pertenece al conjunto, ponemos _tipo = '--'_, lo que significa "línea cortada". Si quisiéramos usar una línea llena (en caso de que la frontera pertenezca al conjunto), usaríamos el valor '-' (como hicimos con la otra parte de la frontera).
  
- La línea `fr2 = frontera(y = '3',tipo = '-')` es análoga a la anterior, pero ahora tenemos una semirrecta horizontal, y entonces $y = f(x)$ (la variable independiente es $x$ y la función es $y$).

- La línea `graficar_region(x_lim,y_lim,cond,fronteras = [fr1,fr2])` es similar a la que usamos en el ejemplo anterior (sin fronteras), pero incluyendo un argumento extra en la función _graficar_region_. Este argumento, que se llama _fronteras_, es una lista que contiene todas las fronteras que definimos anteriormente (separadas por comas, como en cualquier lista).

## Ejercicio 13:

En este ejercicio, se nos da la siguiente función:

$$f(x) = 10 - \frac{1}{200}x$$

Esta función representa la temperatura del aire $y \, [º C]$ a una altura $x\, [m]$ sobre el nivel del mar.

Vamos a programarla en una celda de código:

```python
def temperatura(x): # Defino una función (lineal en este ejemplo)
  a = 10 # Ordenada al origen
  b = -1/200 # Pendiente  
  y = a + b*x # Calculo el valor de la función

  return y # Devuelvo el resultado de la cuenta (temperatura del aire)
```

Como vemos, las funciones se definen en _python_ usando el comando `def` seguido del nombre de la función (en este caso se llama _temperatura_) y de los argumentos de esa función (en este caso únicamente la variable _x_), que se colocan entre paréntesis. Luego, en el cuerpo de la función podemos hacer todas las operaciones que queramos. En este ejemplo primero definimos `a = 10` como la ordenada al origen de la recta (notemos que la función dada tiene la forma $f(x) = a + b.x$, donde $b$ es la pendiente y $a$ es la ordenada al origen; _sí, intercambié las letras que se suelen usar para estos parámetros, pero eso no importa, ¿verdad?_). Luego, definimos la pendiente `b = -1/200`. Luego hicimos la cuenta `y = a + b*x` y la guardamos en la variable llamada _y_. Notemos que en _python_ se usa **\*** para multiplicar, **/** para dividir y **\*\*** para elevar al cuadrado, entre muchas otras operaciones. Finalmente, la línea `return y` **devuelve** el resultado de la cuenta, que está guardado en la variable _y_.

Para usar esta función, luego de ejecutar la celda que la define tenemos que llamarla de la forma `temperatura(x)`, poniendo algún valor en _x_. Veamos ejemplos:

```python
x = 0 # Le asigno algún valor a x
print('f({}) = {} ºC'.format(x,temperatura(x))) # Imprimo el resultado

x = 300 # Le asigno algún valor a x
print('f({}) = {} ºC'.format(x,temperatura(x))) # Imprimo el resultado

x = 1000 # Le asigno algún valor a x
print('f({}) = {} ºC'.format(x,temperatura(x))) # Imprimo el resultado

x = 2000 # Le asigno algún valor a x
print('f({}) = {} ºC'.format(x,temperatura(x))) # Imprimo el resultado
```

Consideremos ahora las preguntas:

b) ¿A qué altura sobre el nivel del mar la temperatura comienza a ser inferior a $0 \, ºC$?
b) ¿A qué altura la temperatura es de $7 \, ºC$?

En este caso, no podemos hacer las cuentas directamente usando nuestra función `temperatura(x)`, porque en estas preguntas los roles están cambiados. Podemos resolver este problema definiendo una nueva función `altura(y)`, que podemos obtener analíticamente despejando $x$ en función de $y$ a partir de la función original:

$$y = 10 - \frac{1}{200}x \implies x = g(y) = 200.(10 - y)$$

Ahora vamos a programar esta nueva función y vamos a calcular lo que necesitamos. Notemos que en la definición de la función (celda siguiente) hacemos todas las cuentas en una sola línea, de forma mucho más compacta que antes (cuando definimos _temperatura(x)_). Sin embargo, en algunos casos conviene hacer las cuentas paso a paso para mejorar la legibilidad del código (otras personas, incluida tu "yo del futuro", podrían querer leerlo y deberían entenderlo fácil).

```python
def altura(y): # Defino una función (lineal en este ejemplo)
  return 200*(10 - y) # Devuelvo el resultado de la cuenta (altura sobre el nivel del mar)

y = 0 # Le asigno algún valor a x
print('g({}) = {} m'.format(y,altura(y))) # Imprimo el resultado

y = 7 # Le asigno algún valor a x
print('g({}) = {} m'.format(y,altura(y))) # Imprimo el resultado
```
