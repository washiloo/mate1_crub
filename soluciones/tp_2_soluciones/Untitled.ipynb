### Ejercicio 2

```python
# a)
fig,ax = plt.subplots(figsize = (8,8))
ax.set_ylim((-2,5))
ax.set_xlim((0,6))
ax.axhline(0,color = 'k',lw = 0.5)
ax.axvline(2,color = 'k',ymax = 6/7)
ax.axvline(5,color = 'k',ymax = 6/7)
ax.axhline(4,ls = '--',color = 'k',xmin = 2/6,xmax = 5/6)
ax.fill_between(x = [2,5],y1 = -2,y2 = 4,alpha = 0.2)
ax.grid()
ax.set_ylabel('y',fontsize = 16)
ax.set_xlabel('x',fontsize = 16)
ax.spines['top'].set_visible(False)
ax.spines['bottom'].set_visible(False)
ax.spines['right'].set_visible(False)
plt.savefig('2a.svg')
```

```python
# b)
fig,ax = plt.subplots(figsize = (8,8))
ax.set_ylim((-2,6))
ax.set_xlim((0,6))
ax.axhline(0,color = 'k',lw = 0.5)
ax.axvline(4,color = 'k',ymax = 7/8)
ax.axhline(5,ls = '--',color = 'k',xmin = 4/6,xmax = 1)
ax.fill_between(x = [4,6],y1 = -2,y2 = 5,alpha = 0.2)
ax.grid()
ax.set_ylabel('y',fontsize = 16)
ax.set_xlabel('x',fontsize = 16)
ax.spines['right'].set_visible(False)
ax.spines['top'].set_visible(False)
ax.spines['bottom'].set_visible(False)
plt.savefig('2b.svg')
```

```python
# c)
fig,ax = plt.subplots(figsize = (8,8))
ax.set_ylim((-2,2))
ax.set_xlim((-2,2))
ax.axhline(0,color = 'k',lw = 0.5)
ax.axvline(0,color = 'k',lw = 0.5)
ax.axhline(0,color = 'k',xmax = 0.5)
ax.axvline(0,color = 'k',ymin = 0.5)
ax.fill_between(x = [-2,0],y1 = 0,y2 = 2,alpha = 0.2)
ax.grid()
ax.set_ylabel('y',fontsize = 16)
ax.set_xlabel('x',fontsize = 16)
ax.spines['right'].set_visible(False)
ax.spines['top'].set_visible(False)
ax.spines['bottom'].set_visible(False)
ax.spines['left'].set_visible(False)
plt.savefig('2c.svg')
```
